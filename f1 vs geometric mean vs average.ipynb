```
import numpy as np
import matplotlib.pyplot as plt
```

# Generate data values for precision and recall

```
p = np.arange(0,1,0.05)
r = np.arange(0,1,0.05)

p,r = np.meshgrid(p,r)
f1 = np.zeros(p.shape)
gm = np.zeros(p.shape)
mean = np.zeros(p.shape)
```

# Compute values for mean, f1 score and geometric mean of precision and recall

```
for i in range(p.shape[0]):
    for j in range(p.shape[1]):
        mean[i,j] = (p[i,j]+r[i,j])/2
        f1[i,j] = 2*p[i,j]*r[i,j]/(p[i,j]+r[i,j]+1e-7) #1e-7 for numerical stability
        gm[i,j] = (p[i,j]*r[i,j])**0.5
```

# Plot results

```
fig = plt.figure(figsize=[18,5])
ax1 = fig.add_subplot(1,3,1, projection='3d')
ax2 = fig.add_subplot(1,3,2, projection = '3d')
ax3 = fig.add_subplot(1,3,3, projection = '3d')
ax1.plot_surface(p,r,mean)
ax2.plot_surface(p,r,gm)
ax3.plot_surface(p,r,f1)
ax1.set_xlabel('precision')
ax1.set_ylabel('recall')
ax2.set_xlabel('precision')
ax2.set_ylabel('recall')
ax3.set_xlabel('precision')
ax3.set_ylabel('recall')
ax1.title.set_text('Mean of precision and recall')
ax2.title.set_text('Geometric mean of precion and recall')
ax3.title.set_text('F1 score (Harmonic mean of precision and recall')
plt.show()
```

#Observations:


*   Arithmetic mean (AM), tend to tradeoff precision/recall over other. Even if precision or recall is very low, AM has slightly higher value than GM or F1 score
*   Geometric Mean(GM), tend to have higher value only when both precision and recall is high.
*   Harmonic Mean(F1 score), tend to have higher value only when both precision and recall is high. But for smaller values of precision or recall Geometric mean has slightly higher value than F1
